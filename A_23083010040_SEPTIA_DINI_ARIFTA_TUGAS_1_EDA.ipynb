VISUALISASI DATA PENJUALAN SETIAP KOTA

```python
import plotly.express as px

# Mengasumsikan 'df' adalah DataFrame asli, Memuat DataFrame 'sales_by_City'
sales_by_City = df.groupby('City', observed=False)['Total'].sum().reset_index()

# Buat grafik batang yang lebih informatif dan menarik
fig = px.bar(
    data_frame=sales_by_City,
    x='City',
    y='Total',
    color='City',
    labels={'Total': 'Total Penjualan', 'City': 'Kota'},
    title='Total Penjualan Berdasarkan Kota',  # Judul grafik
    color_discrete_sequence=px.colors.qualitative.Plotly,  # Skema warna yang bisa diubah sesuai keinginan
    text='Total',  # Menampilkan total penjualan di atas batang
)

# Perbarui layout untuk keterbacaan yang lebih baik
fig.update_layout(
    xaxis_title='Kota',  # Label sumbu x
    yaxis_title='Total Penjualan',  # Label sumbu y
    xaxis_tickangle=-45,  # Memutar label sumbu x agar lebih mudah dibaca
    plot_bgcolor='pink',  # Menetapkan warna latar belakang plot ke putih
    title_x=0.5,  # Menempatkan judul di tengah
)

fig.show()
```

Grafik batang diatas menunjukkan total penjualan untuk tiga kota (Mandalay, Naypyitaw, dan Yangon). Naypyitaw memiliki penjualan tertinggi, diikuti oleh Yangon, dan kemudian Mandalay.

```python
# Group by 'Product line' and calculate the sum of 'Total' for each group
revenue_by_product_line = df.groupby('Product line' , observed=False)['Total'].sum().reset_index()

# Sort the results in descending order to find the highest revenue
revenue_by_product_line = revenue_by_product_line.sort_values(by='Total', ascending=False)

revenue_by_product_line
```

```python
import plotly.express as px

# Buat diagram lingkaran dengan data dari 'revenue_by_product_line'
fig = px.pie(
    revenue_by_product_line,
    names='Product line',
    values='Total',
    title='Pendapatan Berdasarkan Jenis Produk',
    hole=0.3,  # Membuat lubang di tengah untuk efek donat
    color_discrete_sequence=px.colors.qualitative.Plotly  # Menggunakan skema warna cerah
)


fig.update_traces(
    textposition='outside',  # Menempatkan label di luar potongan pie
    textinfo='percent+label',  # Menampilkan persentase dan label
    marker=dict(line=dict(color='#000000', width=1))  # Menambahkan garis tepi hitam pada potongan pie
)


fig.update_layout(
    title_font_size=20,  # Ukuran font judul
    legend_title_text='Jenis Produk',  # Judul legenda
    legend=dict(x=0.8, y=0.5),  # Menempatkan legenda di luar diagram
    paper_bgcolor='pink'  # Latar belakang diagram
)

# Menampilkan diagram
fig.show()
```

Mengetahui Apakah ada bulan-bulan tertentu ketika penjualan mencapai puncaknya?

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Mengatur tema seaborn
sns.set_theme(style="whitegrid")

# Data contoh untuk setiap cabang
data = {
    'Branch A': {
        'transactions': pd.Series([119, 94, 127], index=pd.to_datetime(['2019-01-31', '2019-02-28', '2019-03-31'])),
        'revenue': pd.Series([38681.13, 29860.12, 37659.12], index=pd.to_datetime(['2019-01-31', '2019-02-28', '2019-03-31']))
    },
    'Branch B': {
        'transactions': pd.Series([111, 109, 112], index=pd.to_datetime(['2019-01-31', '2019-02-28', '2019-03-31'])),
        'revenue': pd.Series([37176.06, 34424.27, 34597.34], index=pd.to_datetime(['2019-01-31', '2019-02-28', '2019-03-31']))
    },
    'Branch C': {
        'transactions': pd.Series([122, 100, 106], index=pd.to_datetime(['2019-01-31', '2019-02-28', '2019-03-31'])),
        'revenue': pd.Series([40434.68, 32934.98, 37199.04], index=pd.to_datetime(['2019-01-31', '2019-02-28', '2019-03-31']))
    }
}

# Mengatur ukuran plot
plt.figure(figsize=(15, 10))

# Looping untuk membuat plot per cabang
for i, (branch, branch_data) in enumerate(data.items()):
    # Plot transaksi bulanan
    plt.subplot(2, 3, i + 1)  # 2 baris, 3 kolom, urutan subplot
    sns.lineplot(x=branch_data['transactions'].index.month, y=branch_data['transactions'].values)
    plt.title(f'Monthly Transactions - {branch}')
    plt.xlabel('Month')
    plt.ylabel('Number of Transactions')

    # Plot pendapatan bulanan
    plt.subplot(2, 3, i + 1 + 3)  # Baris kedua untuk pendapatan
    sns.lineplot(x=branch_data['revenue'].index.month, y=branch_data['revenue'].values)
    plt.title(f'Monthly Revenue - {branch}')
    plt.xlabel('Month')
    plt.ylabel('Total Revenue')

# Mengatur tata letak agar tidak overlap
plt.tight_layout()

# Menampilkan plot
plt.show()
```

1. Grafik Transaksi Bulanan (Monthly Transactions):
Sumbu X: Bulan (dari Januari hingga Maret 2019).
Sumbu Y: Jumlah transaksi.
Interpretasi:

Branch A: Terdapat fluktuasi jumlah transaksi dari bulan ke bulan. Misalnya, terjadi penurunan dari Januari ke Februari, tetapi meningkat lagi pada Maret.

Branch B: Jumlah transaksi relatif stabil dengan sedikit penurunan dari Januari ke Februari dan kemudian sedikit kenaikan di Maret.

Branch C: Mengalami fluktuasi yang lebih besar dibandingkan cabang lainnya. Ada penurunan yang cukup signifikan dari Januari ke Februari, tetapi meningkat kembali di Maret.

2. Grafik Pendapatan Bulanan (Monthly Revenue):
Sumbu X: Bulan (dari Januari hingga Maret 2019).
Sumbu Y: Total pendapatan.
Interpretasi:

Branch A: Pendapatan juga mengalami fluktuasi, dengan penurunan dari Januari ke Februari, dan peningkatan kembali pada Maret. Ini mungkin menunjukkan perubahan dalam jumlah transaksi atau nilai transaksi.

Branch B: Ada tren peningkatan pendapatan dari Januari ke Februari, kemudian sedikit penurunan di Maret. Hal ini mungkin mengindikasikan bahwa bulan Februari adalah periode puncak untuk pendapatan.

Branch C: Pendapatan menunjukkan fluktuasi yang cukup signifikan dengan kenaikan dari Januari ke Februari dan sedikit penurunan di Maret. Ini menunjukkan bahwa cabang ini mungkin mengalami perubahan dalam kinerja keuangan atau strategi penetapan harga.

Kesimpulan Umum:
Tren Transaksi dan Pendapatan: Perubahan dalam jumlah transaksi tidak selalu sejalan dengan perubahan pendapatan. Sebagai contoh, jika satu cabang memiliki lebih banyak transaksi tetapi pendapatan lebih rendah, ini mungkin menunjukkan transaksi dengan nilai lebih rendah.

Fluktuasi Bulanan: Perhatikan apakah ada pola musiman atau peristiwa khusus yang dapat mempengaruhi data pada bulan-bulan tertentu. Ini bisa membantu dalam merencanakan strategi bisnis atau memahami faktor-faktor yang mempengaruhi kinerja setiap cabang.
Grafik ini memberikan gambaran yang jelas tentang bagaimana transaksi dan pendapatan berubah dari bulan ke bulan untuk masing-masing cabang, serta memungkinkan perbandingan antar cabang.

MELIHAT PERBANDINGAN DALAM METODE PEMBAYARAN

```python
import pandas as pd

data = {
    'Payment': ['Ewallet']*345 + ['Cash']*344 + ['Credit card']*311
}

df = pd.DataFrame(data)
```

```python
# Menghitung frekuensi setiap metode pembayaran
payment_method_counts = df['Payment'].value_counts().reset_index()
payment_method_counts.columns = ['Payment', 'count']

# Menampilkan hasil
print(payment_method_counts)
```

```python
# Membuat diagram donat
plt.figure(figsize=(8, 8))

# Data untuk plot
sizes = payment_method_counts['count']
labels = payment_method_counts['Payment']
colors = ['#ff9999','#66b3ff','#99ff99']  # Warna untuk setiap bagian

# Plot pie chart dan ubah menjadi donat
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140, wedgeprops=dict(width=0.3))

# Menambahkan judul
plt.title('Payment Method Distribution')

# Menampilkan plot
plt.show()
```
